## `heapq` Tutorial

`heapq` is a priority queue and can be used as the data structure for shortest path search. Here is an example  of using it.

````python
import heapq

priority_queue = [(0, A.get_value(), A)]
cost, value, node = heapq.heappop(priority_queue)
assert cost == 0
assert value == 'A'
assert node == A

# This will pop out B first
heapq.heappush(priority_queue, (1, B.get_value(), B))
heapq.heappush(priority_queue, (2, C.get_value(), C))
cost, value, node = heapq.heappop(priority_queue)
assert cost == 1
assert value == 'B'
assert node == B

# This will pop out C first. When they have same cost, it will compare its value.
heapq.heappush(priority_queue, (2, D.get_value(), D))
cost, value, node = heapq.heappop(priority_queue)
assert cost == 2
assert value == 'C'
assert node == C
````
